La funció **format_sky** us ajudarà a visualitzar el tauler del problema.

````python
import numpy as np
from IPython.display import clear_output
import time

def format_sky(board, top, bottom, left, right):
    """
    Funció auxiliar per mostrar la matriu del Problema 1 en forma de graella.
    """
    _str = ""
    
    # Files
    for v in top:
        _str+="   "+str(v)
    _str = "  " + _str + "\n"

    for idx, i in enumerate(board):
        _str+= "   +"+ ("-"*((board.shape[1]*4)-1)) +"+\n"
        _str+= f" {left[idx]} | "
        
        # Columnes
        for j in i:   
            if j!=0:
                _str+= str(j)+" | "
            else:
                 _str+= "  | "
            
        _str+= f"{right[idx]} \n"
    _str+= "   +"+ ("-"*((board.shape[1]*4)-1))+"+\n"
    
    _bot = ""
    for v in bottom:
        _bot+="   "+str(v)
    _str += "  " + _bot
    
    return _str.replace('0',' ')
````

Veieu un exemple de com utilitzar aquesta funció:


````python
# Amb aquesta funció es mostren totes les caselles del tauler
# Les caselles que contenen un zero es mostren buides

# Creem unes restriccions i un tauler 
top, bottom, left, right = [0,0,0], [3,0,0], [0,0,0], [0,0,1]
grid = np.zeros((len(left), len(top)), dtype='int')

print('Tauler:')
print(format_sky(grid, top, bottom, left, right))
````

La funció **delayed_print** us permet bloquejar temporalment un 'print' per si voleu veure les accions que fa a cada pas el vostre algorisme.<br>
No heu d'utilitzar-la a l'entrega ja que el programa tardarà molt més del necessari, utilitzeu-la només mentre programeu la solució.

````python
def delayed_print(grid, top, bottom, left, right, sleep_time=0.02):
    """
    Funció auxiliar que bloqueja temporalment la funció 'print' per veure l'evolució de l'algorisme.
    """
    print(format_sky(grid, top, bottom, left, right))
    clear_output(wait=True)
    time.sleep(sleep_time)
````

````python
###
# LES VOSTRES FUNCIONS EXTRES AQUÍ
###
from itertools import permutations
from queue import PriorityQueue
###
# LES VOSTRES FUNCIONS EXTRES AQUÍ
###

#Funció per trobar la següent posició buida.
def find_empty(board):
    for i in range(len(board)-1,-1,-1):
        for j in range(len(board)):
            #Considerem un 0 com a posició buida.
            if board[i][j] == 0:
                return (i,j)
    return None

def find_clues_pos(top, right, left, bottom):
    q = PriorityQueue()
    
    if top.count(0) != len(top):
        for i in range(len(top)):
            if top[i] == 1:# or top[i] == 2:
                val = top[i]
                y = i
                x = 0
                pos = (x,y)
                q.put((val, pos))
            if top[i] == len(top):
                val = 0
                x = i
                y = 0
                pos = (x,y)
                q.put((val, pos))
            
    if right.count(0) != len(right):
        for i in range(len(right)):
            if right[i] == 1:# or right[i] == 2:
                val = right[i]
                x = i
                y = len(right)-1
                pos = (x,y)
                q.put((val, pos))
            if right[i] == len(right):
                val = 0
                x = i
                y = len(right)-1
                pos = (x,y)
                q.put((val, pos))
            
    if left.count(0) != len(left):
        for i in range(len(left)):
            if left[i] == 1:# or left[i] == 2: 
                val = left[i]
                x = i
                y = 0
                pos = (x,y)
                q.put((val, pos))
            if left[i] == len(left):
                val = 0
                x = i
                y = 0
                pos = (x,y)
                q.put((val, pos))
        
    if bottom.count(0) != len(bottom):
        for i in range(len(bottom)):
            if bottom[i] == 1 or bottom[i] == 2:
                val = bottom[i]
                y = i
                x = len(bottom)-1
                pos = (x,y)
                q.put((val, pos))
            if bottom[i] == len(bottom):
                val = 0
                y = i
                x = len(bottom)-1
                pos = (x,y)
                q.put((val, pos))
    
    
    return q

#Funció que ens retorna totes les posibles vistes desde el principi de la fila
def possibles_opcions_files_incompletes(fila_param, size):
    posibles_files = []
    nums_totals = list(range(1, size))
    nums_posibles = list(set(nums_totals) - set([x for x in fila_param if x != 0]))
    for perm in permutations(nums_posibles):
        fila = fila_param.copy()
        for e in perm:
            fila[fila.index(0)] = e
        posibles_files.append(fila)
    opcions = set()
    for r in posibles_files:
        opcions.add(opcio_fila(r))
        #Opcions conte una llista amb totes les posibles combinacions d'elements que s'hi poden veure
        #En la llista [0,0,3,0,0] retornaria una llista amb diferents valors segons les permutacions
        #exemple [1,2,3,4,5] afegira 5 a les opcions i [5,4,3,1,2] afegira 1.
    if len(opcions) == 0:
        opcions.add(1)
    return list(opcions)

#Funcio que ens retorna els elements visibles desde la posició on ens situem             
def opcio_fila(fila):
    v = 0
    m = 0
    for x in fila:
        if m < x:
            v += 1
        m = max(m, x)
    #V es el nombre d'elements que podem veure desde el principi de la fila, desde 0 a len(fila)
    return v

def first_steps(grid, x, y, top, bottom, left, right, best_clues, aux):
    if aux == 1:
        if satisfies(grid, x, y, len(top), top, bottom, left, right):
            grid[x][y] = len(top)
            #print(format_sky(grid, top, bottom, left, right))
            #Repetim bactracking
            grid, sol = skyscrapper_backtracking(grid, top, bottom, left, right, best_clues)
            #L'única forma de retornar sol = True, es arribar al punt en que la graella esta plena.
            return grid, True

    if aux == 0:
        if satisfies(grid, x, y, 1, top, bottom, left, right):
            grid[x][y] = 1
            #print(format_sky(grid, top, bottom, left, right))
            grid, sol= skyscrapper_backtracking(grid, top, bottom, left, right, best_clues)
            return grid, True

def satisfies(grid, x, y, num, top, bottom, left, right):
    """
    Donat un tauler, un punt amb coordenades (x,y) i un nombre 'num', comprova si és possible assignar-lo a la posició (x,y) 
    donades les restriccions del problema.
    
    Params
    ------
    :grid: Una matriu de numpy
    :x,y: Un punt de la matriu on volem posar el nombre 'num'
    :num: Un nombre enter
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :b: Un boleà True/False depenent de si el nombre 'num' pot ser col·locat a la casella (x,y) o no.
    """
    
    # EL TEU CODI AQUÍ
    fila = [num if i == y else grid[x][i] for i in range(len(top))]
    col = [num if i == x else grid[i][y] for i in range(len(top))]

    #Comproba si el número està a la fila   
    if fila.count(num) > 1:
        return False

    #Comproba si el número està a la columna
    if col.count(num) > 1:
        return False
    
    cond = top[y]
    if cond != 0 and cond not in possibles_opcions_files_incompletes(col,len(top)+1):
        return False
    
    cond = bottom[y]
    if cond != 0 and cond not in possibles_opcions_files_incompletes(col[::-1],len(bottom)+1):
        return False
            
    cond = right[x]
    if cond != 0 and cond not in possibles_opcions_files_incompletes(fila[::-1],len(right)+1):
        return False
    
    cond = left[x]
    if cond != 0 and cond not in possibles_opcions_files_incompletes(fila,len(left)+1):
        return False
    
    return True
    

def skyscrapper_backtracking(grid, top, bottom, left, right, best_clues):  
    """
    Funció que implementa l'algorisme de backtracking
    
    Params
    ------
    :grid: Una matriu de tipus numpy
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :grid: La matriu plena amb la solució correcta o bé False en cas que no tingui solució.    
    """
    
    # NO esborreu aquestes dues línies, serveien per saber quants cops es crida aquesta funció
    # ----------------------------------------------------------------------------------------
    global num_recursive_calls 
    num_recursive_calls += 1
    # ----------------------------------------------------------------------------------------
    
    # EL TEU CODI AQUÍ   
        
    #Agafem la següent posició buida
    if best_clues.empty():
        pos = find_empty(grid)
        #Si esta plena, retorna la grid
        if not pos:
            return grid, True
        x,y = pos
    else:
        pos = best_clues.get()
        aux, t = pos
        x, y = t
        if not t:
            return grid, True
        first_steps(grid,x,y, top, bottom, left, right, best_clues, aux)
     
    #Hem d'emplenar tants números com caselles, per tant, desde 1 fins al tamany de la graella +1
    for i in range(1, len(top)+1):
        #Comprovem si el número satisfa la condició per ser inclós en aquesta posició
        if satisfies(grid, x, y, i, top, bottom, left, right):
            #Guardem el valor a la posició
            grid[x][y] = i
            #print(format_sky(grid, top, bottom, left, right))
            #Repetim bactracking
            grid, sol = skyscrapper_backtracking(grid, top, bottom, left, right, best_clues)
            #L'única forma de retornar sol = True, es arribar al punt en que la graella esta plena.
            if sol:
                
                return grid, True
            else:
                #Mantenim la posició a 0
                grid[x][y] = 0
    
    return grid, False
        
    
def skyscrapper(top, bottom, left, right):
    """
    Funció principal del problema. Rep quatre llistes corresponents als nombres que hi ha fora del tauler.
    
    Params
    ------
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    """
    q = PriorityQueue()
    
    # Inicialitzem una matriu de zeros
    grid = np.zeros((len(left), len(top)), dtype='int')
    
    q = find_clues_pos(top, right, left, bottom)    
    # Cridem a la funció que soluciona el problema mitjançant backtracking
    grid, sol = skyscrapper_backtracking(grid, top, bottom, left, right,q) 
    
    # Mostrem el resultat en el cas que trobem una solució o mostrem un error en cas contrari.
    if sol:
        print(format_sky(np.array(grid), top, bottom, left, right))
    else:
        print('No solution found')
````

````python
num_recursive_calls = 0
t0 = time.perf_counter()
skyscrapper([0,0,0],
            [3,0,0],
            [0,0,0],
            [0,0,1])
t1 = time.perf_counter()
print('Crides recursives:', num_recursive_calls)
print(f'Temps: {t1-t0:.2f}s')
````
